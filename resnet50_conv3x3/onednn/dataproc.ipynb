```python
%pylab notebook
import pandas as pd
```

```python
df = pd.read_csv('onednn.csv').query('action=="convolution"')
df['shape2'] = df['shape'].str.split(':').str.get(3)
df['size2'] = df['size'].str.extract(r'(oc\d+_ih\d+)')
agg = df.groupby(['size2', 'algo', 'shape2', 'threads']).agg({'time': ['median', 'min', 'max']})
out = agg['time', 'median'].unstack('threads')
out.style.background_gradient(cmap='Greens')
```

```python
df2 = pd.read_csv('gemm.csv')
agg = df2.groupby(['channels', 'imsize', 'version', 'threads']) \
         .agg({'convert': ['median', 'min', 'max'], 'compute': ['median', 'min', 'max']})
out = agg['compute', 'median'].unstack('threads')
out.style.background_gradient(cmap='Greens')
```

```python
out2 = agg['convert', 'median'].unstack('threads')
out2.style.background_gradient(cmap='Greens')
```

```python
df3 = pd.read_csv('sparse.csv')
tmp = df2.query('threads == 40 and version != "sparse.8"').copy()
tmp['sparsity'] = -1
df3 = pd.concat([df3, tmp])
out3 = df3.groupby(['channels', 'imsize', 'version', 'sparsity']) \
          .agg({'convert': 'median', 'compute': 'median'})
out3 = out3['compute'].unstack(['channels', 'imsize'])
out3.style.background_gradient(cmap='Greens', vmin=15, vmax=600)
```
